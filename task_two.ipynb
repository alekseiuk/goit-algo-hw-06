```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
```

```python
G = nx.Graph()
```

```python
stations = [
    "Central", "North Station", "South Station", "East Station",
    "West Station", "University", "Airport", "Industrial Zone",
    "Old Town", "Harbor"
]
```

```python
edges = [
    ("Central", "North Station"),
    ("Central", "South Station"),
    ("Central", "East Station"),
    ("Central", "West Station"),
    ("North Station", "University"),
    ("South Station", "Industrial Zone"),
    ("East Station", "Airport"),
    ("West Station", "Old Town"),
    ("Old Town", "Harbor"),
    ("Harbor", "Industrial Zone")
]
```

```python
G.add_nodes_from(stations)
G.add_edges_from(edges)
```

```python
def dfs_recursive(graph, vertex, visited=None):
    if visited is None:
        visited = set()
    visited.add(vertex)
    print(vertex)  # Відвідуємо вершину
    for neighbor in graph[vertex]:
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited)
```

```python
dfs_recursive(G, "Industrial Zone")
```

```python
def bfs_recursive(graph, queue, visited=None):
    if visited is None:
        visited = set()
    if not queue:
        return
    vertex = queue.popleft()
    if vertex not in visited:
        print(vertex)
        visited.add(vertex)
        queue.extend(set(graph[vertex]) - visited)
    bfs_recursive(graph, queue, visited)
```

```python
bfs_recursive(G, deque(["Industrial Zone"]))
```
