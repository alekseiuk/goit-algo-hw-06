```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
stations = [
    "Central", "North Station", "South Station", "East Station",
    "West Station", "University", "Airport", "Industrial Zone",
    "Old Town", "Harbor"
]
```

```python
edges = [
    ("Central", "North Station"),
    ("Central", "South Station"),
    ("Central", "East Station"),
    ("Central", "West Station"),
    ("North Station", "University"),
    ("South Station", "Industrial Zone"),
    ("East Station", "Airport"),
    ("West Station", "Old Town"),
    ("Old Town", "Harbor"),
    ("Harbor", "Industrial Zone")
]
```

```python
G.add_nodes_from(stations)
G.add_edges_from(edges)
```

```python
# Plot the graph
plt.figure(figsize=(10, 7))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=800)
plt.title("Transportation Network of a Fictional City")
plt.show()
```

```python
# Compute network statistics
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
degree_dict = dict(G.degree())
```

```python
print("Number of nodes:", num_nodes)
```

```python
print("Number of edges:", num_edges)
```

```python
print("Degree of each node:", degree_dict)
```
